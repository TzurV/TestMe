```python
from os import listdir
from os.path import isfile, join

import glob
import re

#https://stackoverflow.com/questions/3207219/how-do-i-list-all-files-of-a-directory

mypath = 'C:\\Users\\tzurv\\Google Drive\\Colab_files\\waves_yesno'

onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f)) and f.endswith('.wav')]

print(onlyfiles)

for f in onlyfiles:
    f1 = re.sub('.wav','',f)
    #print(f1)
    words = ['']
    for l in [char for char in f1 if not char == '_' ]:
        #print(l)
        if l == '0':
            words.append('no')
        else:
            words.append('yes')
            
    spokenWords = ' '.join(words)
    print(f"{f1}: {spokenWords}\n")

#mylist = [f for f in glob.glob(mypath+"\\*.wav")]
#print(mylist)
```
